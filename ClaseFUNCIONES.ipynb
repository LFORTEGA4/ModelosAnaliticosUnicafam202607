```python
from typing_extensions import MutableSequence
#Genera una calculadora de operaciones
def operaciones (a,b):
  suma=a+b
  resta=a-b
  mult=a*b
  division=a/b
  return suma,resta,mult,division
result_suma, result_resta, result_mult, result_division=operaciones(8,5)
print(f"Suma: {result_suma}, Resta:{result_resta}, Multiplicacion:{result_mult}, Division: {result_division}")
```

#Funcion de IMC

```python

def calcular_imc(peso,altura):
  #Calcula el indice de masa corporal
  imc= peso/ (altura ** 2)
  if imc < 18.5:
    return f"IMC:{imc:.2f} - Bajo peso"
  elif 18.5 <= imc <= 24.9:
    return f"IMC:{imc:.2f} - Normal"
  elif 25 <= imc <= 29.9:
    return f"IMC: {imc:.2f} - Sobrepeso"
  else:
    return f"IMC:{imc:.2f} - OBESIDAD"

peso_usuario = float(input("ingresa el peso (en kg): "))
altura_usuario = float(input("ingresa la altura (en metros): "))

resultado_imc = calcular_imc(peso_usuario, altura_usuario)
print(resultado_imc)
```
