Crea un programa que verifique si una persona es mayor de edad. Una persona es considerada mayor de edad si tiene 18 años o más.

```python
# VERIFICAR SI ERES MAYOR DE EDAD
edad = int(input("Introduce tu edad: "))

# Verificamos si es mayor o igual a 18
if edad >= 18:
    print("Eres mayor de edad.")
else:
    print("Eres menor de edad.")
```

Escribe un programa que verifique si un número ingresado por el usuario es par o impar.

```python
# VERIFICAR SI UN NUMERO ES PAR O IMPAR
numero = int(input("Introduce un número: "))
if numero % 2 == 0:
    print("El número es par.")
else:
    print("El número es impar.")
```

Escribe un programa que clasifique la edad en diferentes categorías: Niño (0-12 años), Adolescente (13-17 años), Adulto (18-64 años) y Anciano (65 años o más).

```python
edad = int(input("Introduce tu edad: "))
# Evaluamos las categorías en orden de menor a mayor
if edad < 0:
    print("Edad no válida.")
elif edad <= 12:
    print("Categoría: Niño")
elif edad <= 17:
    print("Categoría: Adolescente")
elif edad <= 64:
    print("Categoría: Adulto")
else:
    print("Categoría: Anciano")
```

Crea un programa que lea dos números y determine cuál es mayor, cuál es menor o si ambos son iguales.

```python
# Solicitamos los dos números al usuario
numero1 = float(input("Introduce el primer número: "))
numero2 = float(input("Introduce el segundo número: "))

# Comparamos los números utilizando condiciones
if numero1 > numero2:
    print(f"El mayor es {numero1} y el menor es {numero2}.")
elif numero1 < numero2:
    print(f"El mayor es {numero2} y el menor es {numero1}.")
else:
    print("Ambos números son iguales.")
```

Escribe un programa que pida al usuario una contraseña y verifique si es correcta. La contraseña correcta es "secreta123".

```python
CONTRASEÑA_CORRECTA = "secreta123"

# Solicitamos al usuario que ingrese la contraseña
intento = input("Introduce la contraseña:")

# Verificamos si el texto ingresado coincide
if intento == CONTRASEÑA_CORRECTA:
    print("Acceso concedido.")
else:
    print("Contraseña incorrecta. Acceso denegado.")
```

Crea un programa que determine si un número ingresado por el usuario está en el rango de 1 a 100 (inclusive) y si es divisible por 5.

```python
numero = int(input("Introduce un número entero: "))

# Verificamos si cumple el rango (1 a 100) Y si es divisible por 5
if numero >= 1 and numero <= 100 and numero % 5 == 0:
    print("El número está en el rango de 1 a 100 y es divisible por 5.")
else:
    print("El número NO cumple con ambas condiciones.")
```
