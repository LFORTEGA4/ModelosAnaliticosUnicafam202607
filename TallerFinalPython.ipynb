# SECCION CONDICIONALES

```python
# 1. Que rama se ejecuta?
puntaje = 72
if puntaje >= 90:
    print("Sobresaliente")
elif puntaje >= 80:
    print("Notable")
elif puntaje >= 70:
    print("Aceptable")
elif puntaje >= 60:
    print("Suficiente")
else:
    print("Reprobado")
```

```python
#2. Clasificador de IMC
peso   = 70
altura = 1.75
imc    = peso / altura ** 2

print(f"IMC: {imc:.1f}")

if imc < 18.5:
    print("Bajo peso")
elif imc < 25:
    print("Normal")
elif imc < 30:
    print("Sobrepeso")
else:
    print("Obesidad")
```

```python
# 3. Condicional con bugs
edad = 20
if edad == 18:
    print("Exactamente 18")
elif edad > 18:
    print("Mayor de 18")
else:
    print("Menor de 18")
```

```python
# 4. Acceso a sistema
edad     = 20
password = "abc123"
mayor    = edad > 18
pass_ok  = len(password) > 8

if mayor and pass_ok:
    print("Acceso permitido")
elif not mayor:
    print("Acceso no permitido: edad insuficiente")
elif not pass_ok:
    print("Acceso no permitido: contraseña muy corta")
```

```python
# 5. Calculadora de descuento
monto = 120000

if monto < 50000:
    descuento = 0
elif monto < 100000:
    descuento = 0.10
elif monto < 200000:
    descuento = 0.20
else:
    descuento = 0.30

ahorro = monto * descuento
total  = monto - ahorro
print(f"Descuento: {int(descuento*100)}% (${ahorro:,.0f})")
print(f"Total: ${total:,.0f}")
```

```python
# 6. FizzBuzz clásico
def fizzbuzz(n):
    if n % 3 == 0 and n % 5 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)

fizzbuzz(15)
fizzbuzz(9)
fizzbuzz(10)
fizzbuzz(7)
```

```python
#1. ¿Que imprime el for?
for i in range(5):
    print(i)
for x in range(2, 10, 3):
    print(x)
for letra in "Hola":
    print(letra)
```

```python
# 2. Suma acumulada
suma_total = 0
for i in range(1, 101):
    suma_total += i
print(f"La suma de todos los números del 1 al 100 es: {suma_total}")
suma_pares = 0
for i in range(2, 101, 2):
    suma_pares += i
print(f"La suma de solo los números pares del 1 al 100 es: {suma_pares}")
```

```python
# 3. Recorrer una lista
precios = [12000, 45000, 8500, 67000, 3200]
caros = 0

for i, p in enumerate(precios, 1):
    print(f"Item {i}: ${p:,}")
    if p > 10000:
        caros += 1

print(f"Artículos > $10,000: {caros}")
```

```python
# 4. While con condición
saldo = 1000
mes   = 0
while saldo > 0:
    saldo -= 150
    saldo -= 20
    mes   += 1
    print(f"Mes {mes}: saldo ${saldo}")

print(f"Saldo agotado en {mes} meses")
```

```python
# 5 break y continue
numeros = [3, 7, -2, 15, 4, -8, 9, 1]
for num in numeros:
    if num < 0:
        continue
    if num > 10:
        print(f"Encontré {num}, deteniendo.") # Fixed indentation and 'n' to 'num'
        break
    print(f"Procesando número: {num}")
```

```python
# 6 Tabla de multiplicar
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i}x{j}={i*j}", end="  ")
    print()
```
