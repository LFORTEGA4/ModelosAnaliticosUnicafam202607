# MENÚ CON 6 OPCIONES

```python
# 1. Diccionario inicial del paciente
paciente = {
    "nombre": "Luisa Ortega",
    "edad": 30,
    "diagnostico": "Asma",
    "medicamentos":["Etinilestradiol"],
    "alergias": ["Metocarbamol"],
    "estado": "En observación"
}
```

```python
# Mostrar las 6 opciones del menú médico
print("=PROCESO DE HISTORIA CLÍNICA =")
print("1. Ver historia clinica completa del paciente")
print("2. Registrar un nuevo medicamento (Agregar par clave-valor)")
print("3. Actualizar edad del paciente (Modificar par clave-valor)")
print("4. Alta médica (Eliminar estado permanentemente con 'del')")
print("5. Derivar paciente (Extraer diagnóstico clínico con '.pop()')")
print("6. Salir del programa")
```

```python
# Capturar la decisión del médico
opcion = input("Seleccione una opción (1-6): ")

# 2. Estructura de control lineal usando solo if / elif / else
if opcion == "1":
    print("\n--- Expediente Clínico Actual ---")
    print(paciente)

elif opcion == "2":
    # Objetivo: Agregar clave-valor por asignación directa
    alergia_detectada = input("Ingrese la alergia detectada: ")
    paciente["alergia_reportada"] = alergia_detectada
    print(f"\nÉxito: Se agregó la clave 'alergia_reportada' con el valor '{alergia_detectada}'.")
    print(paciente)

elif opcion == "3":
    # Objetivo: Actualizar clave-valor por asignación directa
    nueva_edad = int(input("Ingrese la edad actualizada del paciente: "))
    paciente["edad"] = nueva_edad
    print(f"\nÉxito: La edad del paciente se actualizó a {paciente['edad']} años.")

elif opcion == "4":
    # Objetivo: Eliminar datos definitivamente con 'del'
    del paciente["estado"]
    print("\nÉxito: El paciente ha sido dado de alta. Se eliminó la clave 'estado' permanentemente.")
    print(paciente)

elif opcion == "5":
    # Objetivo: Eliminar datos devolviendo el valor con '.pop()'
    diagnostico_archivado = paciente.pop("diagnostico")
    print(f"\nÉxito: El diagnóstico '{diagnostico_archivado}' fue removido del expediente activo.")
    print("El dato extraído fue guardado con éxito en el archivo clínico histórico.")
    print(paciente)

elif opcion == "6":
    print("\nPrograma finalizado correctamente. ¡Tenga un excelente turno, Doctor!")

else:
    print("\nError: Opción inválida. El programa se cerrará. Ejecútelo nuevamente para intentar.")
```

#AGREGAR NUEVO DATO


```python
# 1. Pedir clave y valor al usuario
nueva_clave = input("Ingrese el nombre del nuevo parámetro médico (clave): ").strip()
nuevo_valor = input(f"Ingrese el valor para '{nueva_clave}': ").strip()

# 2. Verificar si la clave ya existe en el diccionario antes de actuar
if nueva_clave in paciente:
    print(f"\n[ALERTA] La clave '{nueva_clave}' YA EXISTE en el expediente.")
    print(f"Valor anterior: {paciente[nueva_clave]}")
    print(f"El sistema procederá a ACTUALIZARLA con el nuevo valor: {nuevo_valor}")
else:
    print(f"\n[INFO] La clave '{nueva_clave}' es nueva. Se AGREGARÁ al expediente.")

# 3. Insertar/Actualizar el par en el diccionario mediante asignación directa
paciente[nueva_clave] = nuevo_valor

# 4. Mostrar el diccionario completamente actualizado
print("\n--- EXPEDIENTE CLÍNICO FINAL ACTUALIZADO ---")
print(paciente)
```

# ACTUALIZACION DE UN DATO EXISTENTE

```python

# 1. Solicitar la clave que se quiere actualizar
clave_a_actualizar = input("Ingrese la clave médica que desea modificar: ").strip()

# 2. Verifica si esa clave ya existe en el diccionario
if clave_a_actualizar in paciente:
    # 3. Si existe, pide el nuevo valor y actualízalo
    print(f"[VERIFICADO] La clave '{clave_a_actualizar}' existe actualmente.")
    print(f"Valor actual: {paciente[clave_a_actualizar]}")

    nuevo_valor = input(f"Ingrese el NUEVO valor para '{clave_a_actualizar}': ").strip()
    paciente[clave_a_actualizar] = nuevo_valor  # Asignación directa

    print(f"\n[OK] Éxito: Se ha actualizado '{clave_a_actualizar}' correctamente.")
else:
    # 4. Si no existe, muestra un mensaje indicándolo
    print(f"\n[ERROR] La clave '{clave_a_actualizar}' NO existe en el expediente.")
    print("No se realizaron cambios para evitar crear parámetros incorrectos.")

# Mostrar el diccionario para verificar el resultado en esta etapa
print("\n--- Estado actual del diccionario ---")
print(paciente)
```

# ELIMINAR UN DATO CON del

```python
# 1. Solicita la clave que se quiere eliminar
clave_a_eliminar = input("Ingrese la clave médica que desea eliminar permanentemente: ").strip()

# 2. Verifica si esa clave existe en el diccionario antes de eliminarla
if clave_a_eliminar in paciente:
    # 3. Si existe, la elimina sin conservar su valor
    del paciente[clave_a_eliminar]
    print(f"\n[OK] Éxito: La clave '{clave_a_eliminar}' ha sido borrada permanentemente.")
else:
    # 4. Si no existe, el programa lo indica amigablemente en vez de detenerse con un error
    print(f"\n[AVISO] La clave '{clave_a_eliminar}' NO existe en el expediente.")
    print("No se realizó ninguna acción. El programa continúa ejecutándose con seguridad.")

# Mostrar el diccionario para verificar el resultado de la eliminación
print("\n--- Estado actual del diccionario ---")
print(paciente)
```

#Eliminar un dato con .pop()

```python
# 1. Solicita la clave que se quiere eliminar
clave_a_extraer = input("Ingrese la clave médica que desea extraer (.pop): ").strip()

# 2. Verifica si la clave existe en el diccionario antes de eliminarla
if clave_a_extraer in paciente:
    # 3. Elimina la clave y guarda el valor en una variable
    valor_eliminado = paciente.pop(clave_a_extraer)

    # 4. Muestra el valor extraído en pantalla
    print(f"\n[OK] Éxito: La clave '{clave_a_extraer}' fue removida del expediente activo.")
    print(f"[HISTORIAL] El valor conservado es: {valor_eliminado}")
else:
    # Mensaje de control si la clave no existe para evitar que el programa falle
    print(f"\n[AVISO] La clave '{clave_a_extraer}' NO existe en este expediente.")
    print("No se pudo realizar la extracción de datos.")

# Mostrar el diccionario para verificar que la clave ya no está
print("\n--- Estado actual del diccionario ---")
print(paciente)
```

# Consultar un dato con .get()

```python
# 1. Solicita la clave que se quiere consultar
clave_a_consultar = input("Ingrese la clave médica que desea consultar: ").strip()

# 2. Busca la clave de forma segura usando .get() con un valor por defecto
# Sintaxis: diccionario.get(clave, valor_si_no_existe)
valor_resultado = paciente.get(clave_a_consultar, "No disponible")

# 3. Muestra el resultado en pantalla
if valor_resultado != "No disponible":
    print(f"\n[OK] Clave encontrada exitosamente.")
    print(f"-> {clave_a_consultar.capitalize()}: {valor_resultado}")
else:
    print(f"\n[AVISO] La clave '{clave_a_consultar}' no existe en este expediente.")
    print(f"-> Resultado: {valor_resultado}")
```
